```python
for column in df:
    print(f'df{column}: {df[column].unique()}')
```

```python
def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes =='object':
            print(f'df{column}: {df[column].unique()}')
```

```python
print_unique_col_values(newdf)
```

```python
newdf.replace('No internet service', 'No', inplace = True)
newdf.replace('No phone service', 'No', inplace= True)
print_unique_col_values(newdf)
```

```python
yes_no_columns = ['Dependents','Partner', 'PhoneService', 'MultipleLines', 'OnlineSecurity' ,'OnlineBackup','DeviceProtection' ,'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']

for col in yes_no_columns:
    newdf[col].replace({'Yes': 1, 'No': 0}, inplace= True)
```

```python
for col in newdf:
    print(f'{col}: {newdf[col].unique()}')
```

```python
newdf['gender'].replace({'Female':1, 'Male':0}, inplace = True)
```

```python
newdf['gender'].unique()
```

```python
#Applied One hot encoding for categorical data columns
```

```python
newdf2 = pd.get_dummies(data= newdf, columns= ['InternetService', 'Contract', 'PaymentMethod'])
newdf2.columns
```

```python
newdf.columns
```

```python
newdf2.sample(5)
```

```python
newdf2.dtypes
```

```python
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
newdf2[cols_to_scale] = scaler.fit_transform(newdf2[cols_to_scale])
```

```python
for col in newdf2:
    print(f'{col}: {newdf2[col].unique()}')
```

```python
# Train Test Split
```

```python
X= newdf2.drop('Churn', axis = 'columns')
y= newdf2['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

```python
# Build a model (ANN) in tensorflow/ keras
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp= model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_pred)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot= True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
